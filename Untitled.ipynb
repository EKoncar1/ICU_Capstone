```python
# our ML algorithm will not be able to understand non-numeric values
# transform categorical to numeric
le = LabelEncoder()
categorical = ['AGE_PERCENTIL', 'WINDOW']
for column in categorical:
    df[column] = le.fit_transform(df[column])
print(df)
```

```python
# Replace missing values with mean 
missing = ['ALBUMIN_MEAN', 'BE_ARTERIAL_MEAN', 'BE_VENOUS_MEAN', 'BIC_ARTERIAL_MEAN', 'BIC_VENOUS_MEAN', 'BILLIRUBIN_MEAN', 'BLAST_MEAN', 'CALCIUM_MEAN', 'CREATININ_MEAN', 'FFA_MEAN', 'GGT_MEAN', 'GLUCOSE_MEAN', 'HEMATOCRITE_MEAN', 'HEMOGLOBIN_MEAN', 'INR_MEAN', 'LACTATE_MEAN', 'LEUKOCYTES_MEAN', 'LINFOCITOS_MEAN', 'NEUTROPHILES_MEAN', 'P02_ARTERIAL_MEAN', 'P02_VENOUS_MEAN', 'PC02_ARTERIAL_MEAN', 'PC02_VENOUS_MEAN', 'PCR_MEAN', 'PH_ARTERIAL_MEAN', 'PH_VENOUS_MEAN', 'PLATELETS_MEAN', 'POTASSIUM_MEAN', 'SAT02_ARTERIAL_MEAN', 'SAT02_VENOUS_MEAN', 'SODIUM_MEAN', 'TGO_MEAN', 'TGP_MEAN', 'TTPA_MEAN', 'UREA_MEAN', 'DIMER_MEAN', 'BLOODPRESSURE_DIASTOLIC_MEAN', 'BLOODPRESSURE_SISTOLIC_MEAN', 'HEART_RATE_MEAN', 'RESPIRATORY_RATE_MEAN', 'TEMPERATURE_MEAN', 'OXYGEN_SATURATION_MEAN']
for col in missing:
     df[col].fillna(df[col].mean(), inplace=True)
df.fillna(df.mean())
```

```python
# Handle infinite values
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(df.mean(), inplace=True)
```

```python
# Immunocompromised Distribution
def immuno(IMMUNOCOMPROMISED):
    if IMMUNOCOMPROMISED == 1:
        return "Patient is Immunocompromised"
    elif IMMUNOCOMPROMISED == 0:
        return "Patient is not Immunocompromised"
    else:
        return "Missing Data"
    
df['immuno']=df['IMMUNOCOMPROMISED'].apply(immuno)
df['immuno'].value_counts().plot(kind='pie', legend=True)
plt.title("Immunocompromised Distribution")
```

```python
# Gender Distribution 
def gender_split(GENDER):
    if GENDER == 0:
        return "Male"
    elif GENDER == 1:
        return "Female"
    else:
        return "Gender Unknown"
    
df['gender_split']=df['GENDER'].apply(gender_split)
df['gender_split'].value_counts().plot(kind="pie", legend=True)
plt.title("Male vs Female Distribution")
plt.xlabel("Gender")
plt.ylabel("Gender Count")
```

```python
# Hypertension Distribution
def htn(HTN):
    if HTN == 0:
        return "Patient does not have hypertension"
    elif HTN == 1:
        return "Patient has hypertension"
    else:
        return "Missing Data"

df['htn']=df['HTN'].apply(htn)
df['htn'].value_counts().plot(kind='pie', legend=True)
plt.title("Hypertension Distribution")
```

```python
# Age Distribution 
sns.countplot(df, x='AGE_PERCENTIL')
plt.title("Age Percentile Distribution")
```

```python
# ICU Admission Status 
def icu_check(ICU):
    if ICU == 1:
        return "Admitted to ICU"
    if ICU == 0:
        return "Not Admitted to ICU"
    else:
        return "Missing Data"
    
df["icu_check"]=df['ICU'].apply(icu_check)
df["icu_check"].value_counts().plot(kind='pie', legend=True)
plt.title("ICU Admission Status")
```

```python
# Define independent and dependent variables
X = ml[['RESPIRATORY_RATE_MEAN', 'BE_VENOUS_MEAN', 'BE_ARTERIAL_MEAN', 'LEUKOCYTES_MEAN', 'UREA_MEAN', 'NEUTROPHILES_MEAN', 'HTN', 'PLATELETS_MEAN', 'POTASSIUM_MEAN', 'BLOODPRESSURE_SISTOLIC_MEAN', 'P02_VENOUS_MEAN', 'TTPA_MEAN', 'DISEASE GROUPING 5', 'DISEASE GROUPING 2']]
y = ml['ICU']
print(X.shape)
print(y.shape)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
# Run logistic regression and calculate accuracy of model 
model = LogisticRegression()
model.fit(X_train, y_train)
model.predict(X_test)
model.score(X_test, y_test)
```

```python
# Tune hyperparameters in order to improve accuracy in Logistic Regression #2
# First, define hyperparameters for logistic regression
hyperparameters = {
    'penalty': ['l1', 'l2'],
    'C': [0.1, 1.0, 10.0]
}

# Create logistic regression model
model = LogisticRegression()

# Use GridSearchCV to tune hyperparameters
gs = GridSearchCV(estimator=model, param_grid=hyperparameters, cv=5)
gs.fit(X_train, y_train)

# Find the best parameters
best_penalty = gs.best_params_['penalty']
best_C = gs.best_params_['C']
best = LogisticRegression(penalty=best_penalty, C=best_C)

# Train the model
best.fit(X_train, y_train)

# Predict on testing set
y_pred = best.predict(X_test)

# Test model accuracy 
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
